# The following is the workflow for news headlines sentiment analysis using nltk-vader_lexicon package

```python
# Importing Required Libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
bitcoin_news_data = pd.read_csv('bitcoin_news_data.csv', 
                                parse_dates=['Date'])
bitcoin_news_data
```

```python
df          = bitcoin_news_data.copy()
df          = df.drop(columns=["Query", "Link", "Summary"])
df['Date']  = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()
```

```python
df
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#intilaize the SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
df['Scores']=df['Title'].apply(lambda Title:sid.polarity_scores(Title))
df['Compound']=df['Scores'].apply(lambda d:d['compound'])
df= df.drop(columns=['Scores', 'Title'])
df
```

```python
grouped_df = df.groupby(df.index).mean()
grouped_df
```

```python
df.to_csv('bitcoin_news_data_sentiment_vader.csv', index=False, encoding='utf-8')
```
