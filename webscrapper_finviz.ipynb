```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
import datetime

finviz_url = "https://finviz.com/quote.ashx?t="

tickers = ['SPY', 'AMZN']

news_tables = {}
for ticker in tickers:
    url = finviz_url + ticker
    req = Request(url=url, headers={'user-agent': 'manish'})
    response = urlopen(req)

    html = BeautifulSoup(response, 'html')
    news_table = html.find(id='news-table')
    news_tables[ticker] = news_table


dict_df_news = {}

for ticker in tickers:
    dict_df_news[ticker] = pd.DataFrame(columns=['date','time','headline'])


for ticker, news_table in news_tables.items():
    
    for row in news_table.findAll('tr'):
        headline = row.a.text
        date_time_ = row.td.text.strip().split(' ')

        if len(date_time_) == 2:
            date = date_time_[0]
            if date.casefold() == 'today':
                date = datetime.date.today().strftime('%b-%d-%y')
            time = date_time_[1]
        else:
            time = date_time_[0]
        
        dict_df_news[ticker] = dict_df_news[ticker].append({'date': date, 'time': time, 'headline': headline}, ignore_index=True)
```
